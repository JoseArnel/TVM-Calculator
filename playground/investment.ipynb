```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
# PV: 50,000, PMT 6000, INT: 7%, FV > 1Mil
data = pd.read_csv('investment.csv')
plt.plot(data.Balance / 10**6)
plt.plot(data.Principal / 10**6)
plt.plot(data.Interest / 10**6)
plt.legend(['Balance', 'Principal', 'Interest'])
plt.xlabel("Year")
plt.ylabel("$(Millions)")
plt.show()
```

```python
plt.plot(data.Principal / 10**6)
plt.plot(data.Interest / 10**6)
plt.legend(['Principal', 'Interest'])
plt.xlabel("Year")
plt.ylabel("$(Millions)")
plt.show()
```
